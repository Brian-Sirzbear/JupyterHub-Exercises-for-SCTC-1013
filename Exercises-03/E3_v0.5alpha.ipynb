## Functions used for Tables:
Just like in Lab 02 Exercises, the bulk of this first part is to provide a reference for the functions you'll use. The tables here are taken from the datascience module, so all these table functions can only work if you import datascience (see the first line of code after this).

Remember that in a given table, lists make up the rows and arrays make up columns. Additionally, tables have separate indexes for rows and columns. An index of 0 could refer to the first column or first row (the header row is not given an index) depending on the function. However, if you're typing `existingTable[index]`, it will go for columns, not rows.

### Functions
|Purpose|Function|
|--|--|
|Creating a New Table (from arrays)|`newTable = Table().with_columns(columnHeader1, array1, columnHeader2, array2, [etc, etc])` (it's much recommended to make a table from |
|Creating a New Table (from .csv file)|`newTable = Table.read_table('fileName')`|
|Revealing the full table|`existingTable.show(numberOfRows)` use empty parentheses if you want all rows|
|Restricting a Table to certain parameters|`existingTable.where(columnHeader, are.condition(parameter))` (check below for conditions)|
|Selecting  row in Table format|`existingTable.take(indexOfRow)`|
|Selecting a column in Table format|`existingTable.select('columnHeader1' OR indexOfColumn1, 'columnHeader2' OR indexOfColumn2, [etc, etc])` OR `existingTable[indexOfColumn]`
|Selecting a column in array format|`existingTable.column('columnHeader1' OR indexOfColumn1, 'columnHeader2' OR indexOfColumn2, [etc, etc])`|
|Adding one row to a Table|`existingTable = existingTable.with_row(newRowAsList1, newRowAsList2, [etc, etc])`|
|Adding one column to a Table|`existingTable = existingTable.with_column(columnHeader1, array1, columnHeader2, array2, [etc, etc])`|
|Adding multiple rows to a Table|`existingTable = existingTable.with_rows(newRowAsList1, newRowAsList2, [etc, etc])`|
|Adding multiple columns to a Table|`existingTable = existingTable.with_columns(columnHeader1, array1, columnHeader2, array2, [etc, etc])`|
|Sorting the rows|`existingTable.sort(columnHeader, descending = True/False)` (descending is automatically set to false; put it to True if you want highest values first)|
|Getting the number of rows/columns|`existingTable.num_rows` OR `existingTable.num_columns`|

### Conditions
|Purpose|Function (in np.where)|
|--|--|
|where target values are equal to the parameter|`existingTable.where(columnHeader, are.equal_to(parameter))` OR `existingTable.where(columnHeader, parameter)`|
|where target values are NOT equal to the parameter|`existingTable.where(columnHeader, are.not_equal_to(parameter))`|
|where target values are less than the parameter|`existingTable.where(columnHeader, are.below(parameter))`|
|where target values are more than the parameter|`existingTable.where(columnHeader, are.above(parameter))`|
|where target values are between two parameters|`existingTable.where(columnHeader, are.between(parameter1, parameter2))` (this includes values equal to parameter1, but NOT those equal to parameter2)|
|where target values contain the parameter|`existingTable.where(columnHeader, are.containing(parameter))`|


One key detail is that you need to keep track of indexes. Taking the index of a table once gives you a column, and taking the index twice gives you a cell. For example, `existingTable[5][2]` goes to the table's 6th column (remember, indexes start counting at 0, not 1) and gets the cell of that column that is in the 3rd row.

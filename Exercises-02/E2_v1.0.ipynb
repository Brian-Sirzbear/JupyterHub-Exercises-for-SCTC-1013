## Lists and Arrays

Lists and Arrays are extremely similar in how they work. Being two of the most-used data types in this program, it is imperative that you get a hold on what is a list vs what is an array, and how they differ in appearance, use, and other aspects.

Note: a component of a list or array is called an element. The placement of an element in a list or array, with the first being designated as 0, the second as 1, and so on, is the Index of the element.

Any and all parts of the code intended to be customizable as befit the situation (e.g. what list are you trying to append stuff to? what variable is this stuff defined as?) has a capital letter (Functions/Methods like these do not use capital letters).

|Composition:|accepts most if not all data types wherever|accepts most of not all data types, but a single array must have only one data type|

### Lists: Element Manipulation
Lists are more data oriented. They typically treat their elements the same regardless of data type. Multiplying by two doesn't double the values of any floats or integers, but instead repeats the composition of the list again.

```python
list1 = [52, 23, 66]
list1*2
```

Concatenation and Mathematical operations: You must specify that you are doing this operation to each element in the list.

```python
list2 = ['hope', 'delight', 'doubt']
[i+'ful' for i in list2]
```

```python
[i*2 for i in list1]
# example of a math operation being done on all elements
```

```python
[len(x) for x in list2]
# example of a function being done on all elements
```

```python
list3 = ['Panama Canal', 'Banana', 'Appalachians']
[x.split('a') for x in list3]
# example of a method being done on all elements
```

The `[x for x in List]` format serves as an operation in itself, and can be done for sublists and subarrays within lists (but it's quite trippy).

```python
list4 = np.arange(1,11) # numbers from 1 to 10
listception = [np.arange(0,x) for x in list4]
listception
```

```python
plus1_listception = [[y+1 for y in x] for x in listception] # that is, add 1 to all the terms for each sublist in the listception
plus1_listception
```

### Arrays: Element Manipulation
It might help to think of arrays as being more numerically oriented. Any mathematical operations on arrays made of floats or integers behave as though the operation is being done on each element individually.

Concatenation and Mathematical Operations: Since Arrays treat numbers like math, you can do any mathematical operations to the array itself, and it will apply to all terms inside.

```python
array1 = make_array(52, 23, 66)
array1*2
```

You can still use the same thing for lists as for arrays, but the output here is a list.

```python
[x**2 for x in array1]
```

Concatenation works slightly differently, but it is still noticeably easier on arrays than on lists.

```python
array2 = make_array('animate', 'corroborate', 'disguise', 'orientate')
array2 = np.char.add(array2,'d')
array2
```

Operations between Arrays is only possible if both have the same number of elements.

```python
array3 = make_array(1,2,3)
array4 = make_array(5,6,7)
array3*array4 # the nth term of array3 is multiplied with the nth term of array4
```

### Arrays as containing only One Datatype
The Numpy Array's main difference from lists is that it limits itself to only one datatype, and any new datatype added to it will cause the array's elements to modify themselves to match a common datatype.

Example: An array made of arrays.

```python
from datascience import *
test1 = make_array('blah', 'blah', 'blah')
test2 = make_array('stuff', 'stuff', 'stuff')
arr = make_array(test1, test2) # array containing two arrays!
print('The First Element here is a', str(type(arr[1])))
print('The Length of this array is', len(arr))
arr
```

If you add another element of a different datatype, the array will accept it, but it will modify its elements such that all of them are the same datatype.

Exhibit 1: arr had arrays previously. If I add a string, the arrays inside are broken apart into their separate strings.

```python
arr = np.append(arr, 'this is a string, not an array.') # adding a string to the array of arrays
print('The First Element here is a', str(type(arr[1])))
print('The Length of this array is', len(arr))
arr
```

Exhibit 2: making an array consisting of a string, an integer, and a float turns all elements into strings. Note that it does this because a string can't reliably be converted into floats or integers the same way floats and integers can be converted into strings.

```python
test3 = make_array('this is a string.', 3, 5.0)
print('The Integer 3 became a', type(test3[1]))
test3
```

## Strings
Strings are one of the three main datatypes alongside floats and integers. While those two are fairly straightforward, the string is more complex because it behaves not like a single character or an inseparable cluster thereof, but rather a sequence in the same way that lists and arrays are sequences of characters.

See the table below for a list of functions.

|Function|String|
|--|--|
|How to get the Element of a certain Index|`existingString[indexOfElement]`|
|How to add Strings together (concatenation)|`newString = existingString+existingString`|
|How to replace elements of a string with another|`existingString = existingString.replace(oldElement, newElement, indexOfoldElement(optional))` (if the index is not given, all instances of the oldElement will be replaced.)|
|How to delete elements from a string|use the above .replace function, but with the newElement being simply an empty quotation|
|How to take a certain region of the string|`newString = existingString.split(boundaryElement)[indexOfTargetSection]` or `newString = existingString[startingIndex:finishingIndex]` (leave the starting or finishing index blank if going from the beginning or going to the end)|
|How to calculate a mathematical expression in String format|`eval(mathExpressionAsString)`|
|How to split up a string|`existingString.split(splitBoundaryAsString)` (this consumes the split boundary and returns a list of the resultant pieces)|

```python
apple = 'apple'
apple = 'h'+apple.replace('p', 'g')
apple
```

```python
One_and_a_half = '2*3/4'
eval(One_and_a_half)
```

## Indices!

Indices are one of the most essential components in Python, because they work with lists, arrays, strings, and (for Lab 03) Tables as well. They are specific integers that represent the position of the terms in the list or array or string. They don't count from 1, but 0 instead. So the 'first' term has an index of 0, the 'second' term has an index of 1, and so forth. See below for the format of these indices.

```python
soups = ['alphabet', 'chicken', 'noodle', 'italian wedding', 'tomato']
soups[0] # this outputs the first term of soups
```

```python
soups[:3] # the colon represents a range; what we're doing is selecting from the start all the way up to (but not including) the term of index 3
```

```python
soups[-1] # going in the negative direction is akin to starting from the last term and working backwards.
```

Strings are also 'lists' of characters, so taking an index gets the nth letter in that string (counting spaces as well). 

```python
word = 'example'
word[0] # indices always start from 0 regardless of the data type.
```

```python
soups[3][1] # the first index gets the term of index 3 ('italian wedding'); the second index gets the letter of index 1 in 'italian wedding'.
```

The integers that make up indices are no different from any other integer, and can be modified so long as it remains an integer.

```python
soups[2+3-1]
```
